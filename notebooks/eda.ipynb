# UAC Care Transition — Exploratory Data Analysis

Time series, KPI trends, and correlation analysis for the UAC care pipeline.

```python
import sys
from pathlib import Path

import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go

ROOT = Path.cwd().parent
sys.path.insert(0, str(ROOT))

from src.preprocessing import load_and_clean
from src.feature_engineering import engineer_features
from src.kpi_engine import compute_correlation_matrix, compute_monthly_aggregates

df = engineer_features(load_and_clean())
df.head()
```

```python
# Time series trends
fig = go.Figure()
for col, color in zip(
    ["apprehensions", "transfers", "hhs_care", "discharges"],
    ["#3b82f6", "#22c55e", "#2563eb", "#a78bfa"],
):
    fig.add_trace(go.Scatter(x=df["date"], y=df[col], name=col, line=dict(color=color)))
fig.update_layout(title="UAC Pipeline Time Series", template="plotly_dark", height=500)
fig.show()
```

```python
# KPI trend analysis
kpi_fig = go.Figure()
for col in ["transfer_efficiency", "discharge_effectiveness", "throughput_rate"]:
    kpi_fig.add_trace(go.Scatter(x=df["date"], y=df[col], name=col))
kpi_fig.update_layout(title="KPI Trends", template="plotly_dark", yaxis_tickformat=".0%", height=400)
kpi_fig.show()
```

```python
# Correlation matrix
corr = compute_correlation_matrix(df)
px.imshow(corr, text_auto=".2f", title="Pipeline Correlation Matrix", color_continuous_scale="Blues").show()
```

```python
# Monthly aggregates
monthly = compute_monthly_aggregates(df)
px.bar(monthly, x="year_month", y=["apprehensions", "transfers", "discharges"], barmode="group",
       title="Monthly Pipeline Volumes").show()
```
